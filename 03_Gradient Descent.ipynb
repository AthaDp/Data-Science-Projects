# Contoh penulisan fungsi matematika

### $f(x) = x^2 + x +1$

```python
def f(x):
    return x**2 + x + 1
```

```python
f(2)
```

```python
#Generate Data
x_1 = np.linspace(start=-3,stop=3,num=100)
```

```python
x_1
```

```python
#plot
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope dan Derivative

```python
def df(x):
    return 2*x + 1
```

```python
df(2)
```

```python
#plot function and derivative side by side

plt.figure(figsize=[15, 5])

plt.subplot(1,2,1)

plt.xlim(-3,3)
plt.ylim(0,8)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1,f(x_1), color='blue', linewidth=3)

plt.subplot(1,2,2)

plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,7)
plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1,df(x_1), color='red', linewidth =5)
plt.show()
```

```python
#plot function and derivative side by side

plt.figure(figsize=[10, 10])

plt.subplot(2,1,1)

plt.xlim(-3,3)
plt.ylim(0,8)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1,f(x_1), color='blue', linewidth=3)

plt.subplot(2,1,2)

plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,7)
plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1,df(x_1), color='red', linewidth =5)
plt.show()
```

## Perulangan Python & Algoritma Gradient Descent 

```python
#Python for loop
for n in range(5):
    print('Hello World', n)
```

```python
#Python while Loop
counter = 0
while counter < 7:
    print('Counting...', counter)
    counter = counter + 1
print("Ready or not here I come!!")
```

```python
#Gradient Descent
new_x = 3
previous_x =0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    #print(step_size)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        break
    
print('Local Minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
print('loop runs for', n, 'times :)')
```

```python
#Superimpose the gradient descent calculations on plot

plt.figure(figsize=[10, 10])

plt.subplot(2,1,1)

plt.xlim(-3,3)
plt.ylim(0,8)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1,f(x_1), color='blue', linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,f(values), color='red', s=100, alpha = 0.6)

plt.subplot(2,1,2)

plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,7)
plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1,df(x_1), color='red', linewidth =5, alpha =0.6)

plt.scatter(x_list,slope_list, color='blue', s=100, alpha = 0.6)

plt.show()
```
